### Question 1

My first question is the following:

> Which team is the best soccer team in each of those 11 country in Europe?

In this question, by using [NetworkX](https://networkx.github.io/) we will make graph  of the teams in different country. And make the number of the game a team wons relative with the size of the node 

Since there are some team with the same short name in those 11 countries so we create a graph of the edges with the id of the teams

```python
G=nx.MultiDiGraph()
G = nx.from_pandas_dataframe(match, 'home_team_api_id', 'away_team_api_id', \
                             ['result','country_name'], create_using=nx.MultiDiGraph())
```

```python
assert len(G.edges()) == len(match)
assert len(G.nodes()) == len(match.home_team_api_id.unique())
```

Make a undirect graph with no multi edges between nodes.

```python
H=nx.Graph(G)
```

Create the width for the edge between edges, this is proportion with the number of game between two team. 
And the size for each node, which proportion with the number of win game of each team

```python
"""Create a list of the width for each edge"""
edgewidth=[]
for (u,v,d) in H.edges(data=True):
    a = 0
    b = 0
    if(G.get_edge_data(v,u)):
        a =  len(G.get_edge_data(v,u))
    if(G.get_edge_data(u,v)):
        b = len(G.get_edge_data(u,v))
    edgewidth.append(a+b)    
```

```python
assert len(edgewidth) == len(H.edges())
assert sum(edgewidth) == len(G.edges())
```

```python
"""Create a dictionary of the size for each node"""
wins=dict.fromkeys(G.nodes(),0.0)
for (u,v,d) in G.edges(data=True):
    r=d['result']
    if r == 'win':
        wins[u]+=1.0
    elif r == 'tie':
        wins[u]+=0.5
        wins[v]+=0.5
    else:
        wins[v]+=1.0
```

```python
assert sum(wins.values()) == len(match)
```

Since we have 11 different countries, so to make it easy, we create a function to return the nodelist and nodesize for each team in a specific country

```python
def get_node_info(G, wins, country_name):
    """Function return the list of nodes and list of node's sizes for a country"""
    nodesize = []
    nodelist = []
    myset = set()
    for (u,v,d) in G.edges(data=True):
        if d['country_name'] == country_name:
            myset.add(u)
            myset.add(v)
    nodelist = list(myset)
    nodesize = [wins[v]*30 for v in nodelist ]
    return (nodelist,nodesize)
```

```python
node_info_list = []
for i in match.country_name.unique():
    node_info_list.append(get_node_info(G, wins, i))
```

```python
assert len(get_node_info(G, wins, 'France')[0]) == \
len(match[match['country_name'] == 'France'].home_team_api_id.unique())
assert sum(get_node_info(G, wins, 'France')[1]) == len(match[match['country_name'] == 'France'])*30
assert len(node_info_list) == 11
```

Create the Graph for 11 countries in which different color for different countried, and label the node with team short name

```python
#from networkx.drawing.nx_agraph import graphviz_layout
#pos=nx.spectral_layout(H)
pos=nx.spring_layout(H,iterations=20)
#pos = graphviz_layout(G, prog='dot')
team_list_label = ['Belgium', 'England', 'France', 'Germany', 'Italy', 'Netherlands',
       'Poland', 'Portugal', 'Scotland', 'Spain', 'Switzerland']
colormap = plt.cm.gist_ncar
color_list = [colormap(i) for i in np.linspace(0, 0.9, 11)]
```

```python
mapping_label = pd.Series(match.home_team.values,index=match.home_team_api_id).to_dict()
```

```python
plt.rcParams['text.usetex'] = False
fig = plt.figure(figsize=(50,50))
nx.draw_networkx_edges(H,pos,alpha=0.3,width=edgewidth, edge_color='m')
nx.draw_networkx_edges(H,pos,alpha=0.4,width=1,edge_color='k')
for i in range(11):
    max_index = node_info_list[i][1].index(max(node_info_list[i][1]))
    max_node = node_info_list[i][0].pop(max_index)
    max_size = node_info_list[i][1].pop(max_index)
    nx.draw_networkx_nodes(H,pos,nodelist=node_info_list[i][0],node_size=node_info_list[i][1],
                        node_color='w'
                       ,alpha=0.5, linewidths = 2.0 )
    nx.draw_networkx_nodes(H,pos,nodelist=[max_node],node_size=[max_size],
                        node_color=color_list[i]
                       ,alpha=0.5, label=team_list_label[i], linewidths = 2.0 )
    
nx.draw_networkx_labels(H,pos,labels = mapping_label, fontsize=14)
plt.legend(numpoints = 1, prop={'size': 50})
font = {'color'      : 'k',
        'fontweight' : 'bold',
        'fontsize'   : 50}
plt.title("Number of win game of soccer teams in Europe", font)
plt.show()
fig.savefig('/data/bnguy123/win_rate.png');
```

Now from the graph we can see that which team in each of eleven countries in Europe had the number of win game from season 2008 to 2016 such as Bayern Munich (BMU) in Germany or Manchester United (MUN) in Enland or Juventus in Italy etc..

### Question 2

My second question is the following:

> Which country has the best average attributes and in which specific attibute that they are leading?

Now let look at the data:

```python
team_att.head()
```

First create a new column of total attribute name total_att and make a point graph of average total attribute of all teams in each coutry (country_name) to get an overview about which country has the best attribute in Europe.

```python
team_att['total_att'] = team_att.loc[:, 'speed':'defence'].sum(1)
alt.Chart(team_att).mark_point().encode(    
    alt.X('country_name', sort=alt.SortField(field='total_att', op='mean', order='ascending')),
    alt.Y('average(total_att):Q'),
)
```

From the graph we can see that Italy is the country has the best attribute. Now we want to see why is this, in which specific attribute Italy is better than the other country. 
First, we group the data by country_name and find the average in each attribute and then melting the dataset to set the attribute columns to variables, so we can make the bar chart of different attributes for each country.

```python
groups = team_att.groupby('country_name')['total_att', 'speed', 'passing', 
                                          'crossing','shooting','defence' ].mean().reset_index()
att = pd.melt(groups, id_vars=['country_name','total_att'], value_vars=['speed', 'passing',
      'crossing','shooting', 'defence'],var_name='attribute', value_name='score')
att.head()
```

Let make the bar chart

```python
alt.Chart(att).mark_bar().encode(
    alt.Y('country_name', sort=alt.SortField(field='total_att', op='average', order='descending')),
    alt.X('sum(score):Q'),
    alt.Color('attribute')
)
```

Now we can see why Italy is the country with the best attribute, it has good score in all the attribute, we can also make a radar char by using plotly to get more information about which specific attribute Italy is leading.

```python
plot_data = groups.drop('total_att', axis = 1)
plot_data.set_index('country_name', inplace=True)
radar_data = list()
for i in range(11):
    radar_data.append(go.Scatterpolar(
          r = plot_data.values[i],
          theta = plot_data.columns.values,
          fill = 'toself',
          name = plot_data.index[i]))


layout = go.Layout(
  polar = dict(
    radialaxis = dict(
      visible = False,
      range = [0, 65]
    )
  ),
  showlegend = True
)

fig = go.Figure(data=radar_data, layout=layout)
py.iplot(fig)
```

Speed and shooting is two attributes that Italy is leading compare to other countries. Another interesting thing we can notice here is that although Netherlands has lowest total attribute, it's passing is highest in Europe.
